```python
import random

print("Welcome to my enhanced game!")
name = input("What is your name? ")
age = int(input("What is your age? "))
print("Hello", name, "you are", age, "years old.")
health = 10
weapon = ''  # Initialize weapon
print("You are starting with", health, "health")

is_older = age >= 18
if is_older:
    print("You are old enough to play!")
    
    wants_to_play = input("Do you want to play? ").lower()
    if wants_to_play == "yes":
        print("Let's play!")

        # Choose a weapon
        weapon = input("Choose your weapon: Sword, Bow, or Wand? ").lower()

        left_or_right = input("First choice... Left or Right (left/right)? ")
        if left_or_right == 'left':
            ans = input("Nice, you follow a path and reach a lake. Do you swim across or go around (across/around)? ")

            if ans == "around":
                print("You went around and reached the other side of the lake.")
            elif ans == "across":
                print("You managed to get across, but were bitten by a fish and lost 5 health.")
                health -= 5

            ans = input("You notice a house and a river. Which do you go to (river/house)? ")
            if ans == 'house':
                print("You go to the house and are greeted by the owner. He doesn't like you, and you lose 5 health.")
                health -= 5

                if health <= 0:
                    print("You now have 0 health, and you lost the game.")
                else:
                    print("You have survived...")

                    # Combat scenario
                    print("As you continue your journey, you encounter a fearsome monster!")

                    # Determine the strength of the monster
                    monster_strength = random.randint(1, 10)
                    print("The monster's strength is:", monster_strength)

                    # Determine if you can defeat the monster based on your weapon and strength
                    if (weapon == 'sword' and monster_strength <= 5) or \
                            (weapon == 'bow' and monster_strength <= 7) or \
                            (weapon == 'wand' and monster_strength <= 8):
                        print("You see the monster approaching from the right. What will you do?")
                        action = input("Type 'attack' to attack or 'save' to try and save yourself: ").lower()

                        if action == 'attack':
                            print("You bravely attack the monster with your", weapon, "and defeat it!")
                            # Add something valuable to the player's inventory or increase health, etc.
                        elif action == 'save':
                            print("You wisely choose to save yourself. You manage to escape the monster.")
                        else:
                            print("Invalid choice. The monster catches you off guard. You lose 3 health.")
                            health -= 3

                            if health <= 0:
                                print("You now have 0 health, and you lost the game.")
                            else:
                                print("You managed to escape the monster, but your health is now", health)

                    else:
                        print("The monster overpowered you. You lose 3 health.")
                        health -= 3

                        if health <= 0:
                            print("You now have 0 health, and you lost the game.")
                        else:
                            print("You managed to escape the monster, but your health is now", health)

        elif left_or_right == 'right':
            print("Suddenly, a monster attacks from the right!")

            # Determine the strength of the monster
            monster_strength = random.randint(1, 10)
            print("The monster's strength is:", monster_strength)

            # Determine if you can defeat the monster based on your weapon and strength
            if (weapon == 'sword' and monster_strength <= 5) or \
                    (weapon == 'bow' and monster_strength <= 7) or \
                    (weapon == 'wand' and monster_strength <= 8):
                print("You see the monster approaching. What will you do?")
                action = input("Type 'attack' to attack or 'save' to try and save yourself: ").lower()

                if action == 'attack':
                    print("You bravely attack the monster with your", weapon, "and defeat it!")
                    # Add something valuable to the player's inventory or increase health, etc.
                elif action == 'save':
                    print("You wisely choose to save yourself. You manage to escape the monster.")
                else:
                    print("Invalid choice. The monster catches you off guard. You lose 3 health.")
                    health -= 3

                    if health <= 0:
                        print("You now have 0 health, and you lost the game.")
                    else:
                        print("You managed to escape the monster, but your health is now", health)

            else:
                print("The monster overpowered you. You lose 3 health.")
                health -= 3

                if health <= 0:
                    print("You now have 0 health, and you lost the game.")
                else:
                    print("You managed to escape the monster, but your health is now", health)

        else:
            print("You fell down and lost...")

    else:
        print("See you later...")
else:
    print("You are not old enough to play!")

# Display the chosen weapon at the end
if weapon:
    print("Thanks for playing, {}, with your trusty {}!".format(name, weapon))
```
